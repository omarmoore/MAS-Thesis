```python
#merging tables 
metadata_test = pd.read_csv('FilmData/movies_metadata.csv')
ratings_test = pd.read_csv('FilmData/ratings.csv')

# Rename the 'id' column to 'movieId' in the metadata table
metadata_test = metadata_test.rename(columns={'id': 'movieId'})

ratings_test['movieId'] = ratings_test['movieId'].astype(str)

# Merge the ratings and metadata tables on 'movieId'
merged_df = ratings_test.merge(metadata_test, how='left', on='movieId')
```

```python
merged_df.head()
```

```python
merged_df["title"].nunique()
```

```python
merged_df["title"].value_counts().head()
```

```python
# Let's get to the movies with less than 1000 reviews:
comment_counts = pd.DataFrame(merged_df["title"].value_counts())
rare_movies = comment_counts[comment_counts["title"] <= 1000].index

# Let's get access to movies with over 1000 reviews:
common_movies = merged_df[~merged_df["title"].isin(rare_movies)]
common_movies.shape
```

```python
common_movies.head()
```

```python
common_movies["title"].nunique()
```

```python
#making a user table
user_movie_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
```

```python
user_movie_df.head(25)
```

```python
#first recommendations base on item features
movie_name = "2001: A Space Odyssey"
movie_name = user_movie_df[movie_name]
user_movie_df.corrwith(movie_name).sort_values(ascending=False).head(10)
```

```python
# Let's determine the movies that the user watched.

# Let's choose random user:
# random_user = int(pd.Series(user_movie_df.index).sample(1, random_state=45).values)
random_user = 12
```

```python
random_user_df = user_movie_df[user_movie_df.index == random_user]
random_user_df
```

```python
movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
movies_watched
```

```python
#let's verify:
user_movie_df.loc[user_movie_df.index == random_user, user_movie_df.columns == "Godzilla"]
# gave this movie a 3.0 rating.
```

```python
len(movies_watched)
```

```python
movies_watched_df = user_movie_df[movies_watched]
movies_watched_df.head()
movies_watched_df.shape
```

```python
user_movie_count = movies_watched_df.T.notnull().sum()

user_movie_count = user_movie_count.reset_index()
user_movie_count.columns = ["userId","movie_count"]
user_movie_count.head()
```

```python
user_movie_count[user_movie_count["movie_count"] == 130].count()
```

```python
perc = len(movies_watched) * 60 / 100
perc
```

```python
users_same_movies = user_movie_count[user_movie_count["movie_count"] > perc]["userId"]
users_same_movies.count()
```

```python
final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movies)],
                      random_user_df[movies_watched]])

final_df.shape
final_df.T.corr()
```

```python
#corr for all users:
corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.index.names = ['user_id_1', 'user_id_2']
corr_df = corr_df.reset_index()
corr_df.head()
```

```python
# Users with a correlation of %65 or more with 28491 users:
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.65)][
    ["user_id_2", "corr"]].reset_index(drop=True)

top_users = top_users.sort_values(by='corr', ascending=False)
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
top_users
```

```python
# let's see the ratings of users:
rating = pd.read_csv('FilmData/ratings.csv')
top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how='inner')

top_users_ratings = top_users_ratings[top_users_ratings["userId"] != random_user]
top_users_ratings.head()
```

```python
# Calculate the Weighted Average Recommendation Score and keep the first 5 movies.

#Let's do a single score with the most similar by corr * rating:
top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})

recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
recommendation_df.head()
```

```python
# weighted rating greater than 4:
recommendation_df[recommendation_df["weighted_rating"] > 3.7]

# Movies 28491 will like:
movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3.7].sort_values("weighted_rating", ascending=False)

movies_to_be_recommend.merge(common_movies[["movieId", "title"]])

#Let's see the top 5 movies:
movies_to_be_recommend.merge(common_movies[["movieId", "title"]])[:5]
```
